```python
'''
     1. What is the difference between interpreted and compiled language? 
           Ans:- The key   
           difference between interpreted and compiled programming languages lies in how they execute code. Interpreted languages execute code directly, line by line, without a separate compilation step. Compiled languages, on the other hand, require a separate compilation phase to translate the code into machine code before execution. 
    
    
    2. What is exception handling in Python? 
    Ans:- Exception handling in Python is a mechanism to gracefully manage errors that occur during the execution of a program. These errors, known as exceptions, can disrupt the normal flow of execution if not handled. Python's exception handling uses try, except, else, and finally blocks to manage errors.
    The try block encloses the code that might raise an exception.
    If an exception occurs within the try block, the program immediately jumps to the corresponding except block.
    The except block specifies how to handle the exception, allowing the program to continue running instead of crashing.
    The optional else block executes if no exceptions were raised in the try block.
    The optional finally block executes regardless of whether an exception was raised or caught, and it is often used for cleanup actions.
    
    3. What is the purpose of finally block in exception handling in Python? 
    Ans:- The finally block in Python's exception handling is designed to execute code that should always run, regardless of whether an exception occurred or not. It provides a mechanism for cleanup actions and resource management, ensuring that essential operations are performed reliably.
    The primary purposes of the finally block are:
    Resource Cleanup:
    It's commonly used to release resources like file handles, network connections, or database cursors. This prevents resource leaks and ensures proper system operation.
    Guaranteed Execution:
    The code within the finally block is always executed, whether an exception was raised and handled, an exception was raised but not handled, or no exception occurred at all.
    Maintaining Program State:
    It helps maintain a consistent program state by performing necessary actions before the program exits the try...except block.
    
    4. What is logging in Python?
    Ans:-
           Python logging is a module that allows you to track events that occur while your program is running. You can use logging to record information about errors, warnings, and other events that occur during program execution. And logging is a useful tool for debugging, troubleshooting, and monitoring your program.
    
    5. What is the significance of __del__ method in Python?
    Ans:-
           The __del__ method in Python, also known as a destructor, is called when an object is garbage collected, which occurs after all references to the object have been deleted. Its primary significance lies in releasing external resources held by an object, such as closing files or network connections, before the object is destroyed. However, the __del__ method is not always guaranteed to be called, and its execution is unpredictable, as it depends on the garbage collector's behavior. Therefore, it is generally recommended to use context managers or explicit cleanup methods instead of relying on __del__ for resource management. 
    
    6. What is the difference between import and from ...import python ?
       Ans:-
            The import statement and the from ... import statement in Python serve different purposes in how they bring modules and their contents into your code.
    import module_name:
    This statement imports the entire module. To access items within the module, you use the module name as a prefix (e.g., module_name.item). It keeps the namespace clean and avoids potential naming conflicts.
    from module_name import item:
    This statement imports specific items (functions, classes, variables) directly from the module into the current namespace. You can use these items without the module name prefix (e.g., item). It can be more convenient but may lead to naming conflicts if multiple modules define items with the same name.
    
    7. How can you handle multiple exceptions in Python?
    Ans:-
           Multiple exceptions in Python can be handled using several approaches:
    Separate except blocks: This involves writing individual except blocks for each exception type. It allows for specific handling of different errors.
    E.g.
    try:
            # Code that might raise exceptions
            result = 10 / 0
        except ZeroDivisionError:
            print("Cannot divide by zero")
        except TypeError:
            print("Type error occurred")
    
    8. What is the purpose of with statement and handling file in Python? 
    Ans:-
            The with statement in Python serves to streamline resource management, particularly in file handling, by ensuring proper setup and teardown processes. It automatically handles the opening and closing of files, even if errors occur within the code block. This eliminates the need for manual resource management using try...finally blocks, resulting in cleaner and more readable code.
    When dealing with files, the with statement guarantees that the file is properly closed after its operations are completed. This is crucial to prevent resource leaks and data corruption, especially in scenarios involving multiple file interactions or potential exceptions.
    Here's an example of using the with statement for file handling:
    
    with open("my_file.txt", "r") as file:
        content = file.read()
        # Process the file content
    # File is automatically closed here
    
    9. What is the difference between multi threading and multi processing?
    Ans:-
           Multithreading and multiprocessing are both techniques to run multiple tasks concurrently, but they differ in how they achieve this. Multithreading creates multiple threads within a single process, allowing for concurrent execution within that process, while multiprocessing creates multiple processes, each with its own resources, enabling parallel execution across multiple processors. 
    
    10. What are the advantages of using logging in a program ?
    Ans:-
           Logging offers significant advantages in software development, including improved debugging, easier troubleshooting, enhanced system observability, and better communication between developers and administrators. It provides a record of events, helping identify issues, understand system behavior, and optimize performance.
        11. What is memory management in Python? 
       Ans:- Memory management in Python involves the allocation and deallocation of memory for objects. Python automatically manages memory through a private heap, which stores all Python objects and data structures. The Python memory manager handles this private heap, utilizing components for sharing, segmentation, preallocation, and caching.
    At a low level, a raw memory allocator ensures sufficient space in the private heap by interacting with the operating system's memory manager. Object-specific allocators operate on the same heap, implementing memory management policies tailored to object types like integers, strings, tuples, and dictionaries. 
    Python uses automatic garbage collection, primarily employing reference counting to track object references and a cyclic garbage collector to handle circular references. When an object's reference count drops to zero, it becomes eligible for garbage collection, and its memory is freed.
    
    12. What are the basic steps involved in exception handling in Python? 
    Ans:-
           Exception handling in Python involves a structured approach to manage errors that may arise during the execution of a program. The basic steps are outlined below:
    Try Block: Enclose the code that might raise an exception within a try block. This signifies the section of code where potential errors are anticipated.
        try:
            # Code that might raise an exception
            result = 10 / 0
        except ZeroDivisionError:
            # Code to handle the exception
            print("Cannot divide by zero.")
    
    
    Except Block(s): Follow the try block with one or more except blocks. Each except block specifies the type of exception it can handle. If an exception occurs in the try block, Python will search for a matching except block to execute.
    
        try:
            value = int("abc")
        except ValueError:
            print("Invalid input. Please enter a number.")
        except TypeError:
           print("Type error occurred")
    
    Else Block (Optional): An optional else block can be included after the except block(s). The code within the else block will execute only if no exceptions were raised in the try block.
    
        try:
            num1 = int(input("Enter a number: "))
            num2 = int(input("Enter another number: "))
            result = num1 / num2
        except ZeroDivisionError:
            print("Cannot divide by zero.")
        except ValueError:
            print("Invalid input. Please enter a number.")
        else:
            print("The result is:", result)
    
    Finally Block (Optional): An optional finally block can be included after the except and else blocks. The code within the finally block will always execute, regardless of whether an exception was raised or caught. It is commonly used for cleanup actions.
    
    
        try:
            file = open("my_file.txt", "r")
            content = file.read()
        except FileNotFoundError:
            print("File not found.")
        finally:
            file.close() # Ensure the file is closed, regardless of exceptions
    
    
    Raise: The raise keyword is used to explicitly trigger an exception. This is useful in situations where you need to signal an error condition based on program logic.
    
        def process_data(data):
            if not isinstance(data, list):
                raise TypeError("Data must be a list.")
            # Further processing of data
    
    
    These steps provide a structured way to handle exceptions, preventing program crashes and allowing for graceful error management.
    
    
    13. Why is memory management important in Python? 
    Ans:-
            Memory management is crucial in Python because it ensures efficient and reliable program execution. By effectively allocating and deallocating memory, Python programs can avoid issues like memory leaks, improve performance, and prevent crashes. 
    Here's why memory management is important in Python:
    Preventing Memory Leaks:
    Memory leaks occur when a program allocates memory but fails to release it when it's no longer needed, leading to a gradual increase in memory usage over time. This can cause performance problems and even crashes if the program runs out of memory. 
    Improving Performance:
    Efficient memory management allows Python programs to run faster and more efficiently by minimizing the amount of memory they require. This can lead to faster processing speeds and reduced resource consumption. 
    Avoiding Crashes:
    By properly managing memory, Python programs can avoid memory exhaustion, which can lead to crashes or unexpected errors. Memory management ensures that programs have sufficient memory to operate without running out of resources. 
    Optimizing Resource Usage:
    Efficient memory allocation and deallocation help optimize resource usage, allowing programs to run smoothly even on systems with limited memory. 
    Supporting Large Data and Applications:
    When dealing with large datasets or complex applications, efficient memory management is essential for preventing performance bottlenecks and ensuring that programs can handle the workload effectively. 
    Ensuring Data Integrity:
    Memory management helps maintain data integrity by preventing memory corruption and ensuring that data is stored and retrieved correctly. 
    
    14. What is the role of try and except block in exception handling? 
    Ans:- 
           The try and except blocks in Python are fundamental to exception handling, allowing programs to gracefully manage errors that would otherwise cause them to crash. The try block encloses the code that might raise an exception. If an exception occurs within this block, the normal flow of execution is interrupted, and the program immediately jumps to the except block. The except block specifies how to handle the exception. It defines the actions to be taken when a particular type of error occurs. If no exception occurs in the try block, the except block is skipped. 
    
    try:
        # Code that might raise an exception
        result = 10 / 0
    except ZeroDivisionError:
        # Handle the exception
        print("Cannot divide by zero.")
    
    15. How does python's garbage collection system works?
    Ans:- 
            , garbage collection is a memory management mechanism that automatically frees up memory occupied by objects that are no longer in use. It's primarily done through a combination of reference counting and generational garbage collection. Reference counting determines if an object is still needed by tracking how many other objects refer to it; when an object's reference count reaches zero, it's deallocated. Additionally, Python uses a generational garbage collector to identify and collect objects that are no longer reachable by the program, ensuring efficient memory usage. 
    Here's a more detailed breakdown:
    1. Reference Counting:
    Each object in Python has a reference count, which indicates how many other objects or parts of the program are referencing it. 
    When a reference to an object is created (e.g., assigning a variable to an object), the reference count increases. 
    When a reference is removed (e.g., the variable is no longer assigned to the object), the reference count decreases. 
    If an object's reference count drops to zero, it means no part of the program is actively using it, and the garbage collector can safely deallocate the memory occupied by that object. 
    2. Generational Garbage Collection:
    Python also uses generational garbage collection, which divides objects into generations based on how long they have survived previous collection cycles.
    New objects are typically placed in the youngest generation (generation 0).
    If an object survives a collection cycle, it might be moved to a higher generation (generation 1 or 2).
    The garbage collector periodically runs and attempts to collect objects in different generations.
    This approach helps to optimize the garbage collection process by focusing on more likely garbage objects in younger generations and leaving older, more stable objects alone. 
    3. Mark-and-Sweep (for generational garbage collection):
    The garbage collector uses a mark-and-sweep algorithm to identify and collect unreachable objects within the generational context.
    Mark: The algorithm starts from a set of root objects (like global variables and local variables) and marks all objects that are directly or indirectly reachable from those root objects.
    Sweep: It then sweeps through all objects in a generation and deallocates those that were not marked as reachable. These unmarked objects are considered garbage. 
    In essence, Python's garbage collection mechanism ensures that memory is automatically freed up when it's no longer needed, preventing memory leaks and allowing for efficient memory usage. 
    
    16. What is the purpose of else block in exception handling? 
       Ans:-
            The else block in Python's exception handling is used to define a block of code that will be executed if no exceptions are raised within the try block. It provides a way to separate the code that might raise an exception from the code that should only run if the try block executes successfully. This enhances code clarity and avoids accidentally catching exceptions that were not intended to be handled by the try-except block.
           try:
        result = 10 / 2
    except ZeroDivisionError:
        print("Cannot divide by zero")
    else:
        print("Division successful. Result:", result)
    
    17. What are the common logging levels in Python?
     Ans:-
           Python's logging module defines five standard logging levels, each serving a different purpose and indicating the severity of a log message:
    DEBUG (10):
    Detailed information, typically used for diagnosing problems. It's helpful during development and testing.
    INFO (20):
    Confirmation that things are working as expected. It provides general operational information.
    WARNING (30):
    Indicates a potential issue or something unexpected that might cause problems in the future, but the application can usually continue.
    ERROR (40):
    Signifies a more serious problem where a function failed to execute, but the application as a whole can likely continue running.
    CRITICAL (50):
    The most severe level, indicating a critical error that might cause the application to terminate.
    These levels are hierarchical, meaning that setting a logging level will capture all messages at that level and above. For example, setting the level to INFO will include INFO, WARNING, ERROR, and CRITICAL messages, but exclude DEBUG messages.
    
    18. What is the basic difference between OS:fork() and multi processing in Python?
    Ans:- 
            The key differences between os.fork() and the multiprocessing module in Python lie in their portability, how they handle resources, and their typical use cases.
    Portability:
    os.fork() is a Unix-specific system call and is not available on Windows. The multiprocessing module is designed to work cross-platform, including Windows.
    Resource Handling:
    os.fork() creates a near-identical copy of the current process, including its memory space, open file descriptors, and other resources. This can be efficient but can also lead to issues if not managed carefully, especially with locks and shared resources. The multiprocessing module, by default, creates new processes with separate memory spaces. This avoids many of the resource management issues of fork() but can introduce overhead for data sharing between processes.
    Library Overhead:
    os.fork() is a low-level system call with minimal overhead. The multiprocessing module is a higher-level interface and involves more setup and management, which can result in higher overhead.
    Use Cases:
    os.fork() is often used for tasks where a process needs to quickly create a copy of itself, often to handle incoming requests or perform parallel operations. The multiprocessing module is better suited for complex parallel processing tasks where processes need to communicate and share data in a controlled manner.
    
    19. What is the importance of closing a file in Python? 
    Ans:- 
            Closing a file in Python is important for several reasons:
    Resource Management:
    When a file is opened, the operating system allocates resources to manage it. Failing to close the file after use can lead to resource leaks, potentially causing performance issues or crashes if many files are left open. 
    Data Integrity:
    Data written to a file might be buffered and not immediately written to the disk. Closing the file ensures that all remaining data in the buffer is flushed and written, preventing data loss or corruption.
    File Locking:
    Some file operations require exclusive access to the file. If a file is not closed, it may remain locked, preventing other processes or users from accessing it. 
    Code Maintainability:
    Explicitly closing files makes the code more readable and maintainable, clearly indicating when file resources are being managed.
    Prevention of Errors:
    Not closing files can sometimes lead to errors, especially when dealing with multiple files or in scenarios with frequent file operations.
    Python's with statement provides a convenient way to automatically close files, even if exceptions occur:
    
    try:
        with open("example.txt", "w") as file:
            file.write("This is some example text.")
    except IOError as e:
        print(f"An error occurred: {e}")
    
        20. What is the difference between file. read ()and file.readline() function in Python? 
    Ans:- The file.read() and file.readline() methods in Python serve distinct purposes when reading data from files:
    file.read([size]):
    This method reads the entire file content, or up to size characters if specified, and returns it as a single string. If size is omitted or negative, the entire file is read. It's suitable for smaller files or when the entire content needs to be processed at once.
    file.readline([size]):
    This method reads a single line from the file, including the newline character (\n), or up to size characters if specified, and returns it as a string. Subsequent calls to readline() will read subsequent lines. It is efficient for processing large files line by line without loading the entire file into memory.
    
    21. What is the logging module in Python used for? 
         Ans:- Python logging is a module that allows you to track events that occur while your program is running. You can use logging to record information about errors, warnings, and other events that occur during program execution. And logging is a useful tool for debugging, troubleshooting, and monitoring your program.
    
    22. What is the OS module in Python is used for in file handling?
    Ans:- Python has a built-in os module with methods for interacting with the operating system, like creating files and directories, management of files and directories, input, output, environment variables, process management, etc.
    
    23. What are the challenges associated with memory management in Python? 
    Ans:- 
          Python's automatic memory management, while simplifying development, presents several challenges:
    Memory leaks:
    Although Python uses garbage collection, memory leaks can still occur, particularly due to reference cycles where objects refer to each other, preventing their reference counts from reaching zero and thus not being deallocated.
    High memory consumption:
    Python's dynamic typing and object-oriented nature can lead to higher memory usage compared to languages like C or C++. This can be a concern when working with large datasets or long-running applications.
    Garbage collection overhead:
    The garbage collector reclaims unused memory, but it can introduce performance overhead, especially with large amounts of data or frequent allocations and deallocations. The timing of garbage collection cycles can also be unpredictable, leading to latency spikes.
    Fragmentation:
    Dynamic allocation and deallocation can lead to memory fragmentation, where free memory is scattered and unusable, even if enough total memory is available. Python's memory manager attempts to mitigate fragmentation, but it can still be an issue.
    Limited manual control:
    Python's memory management is largely automatic, offering limited manual control compared to languages like C or C++. This can make it difficult to optimize memory usage in specific scenarios.
    Multithreading and multiprocessing:
    When using multithreading, race conditions, deadlocks, and synchronization issues can arise. Multiprocessing, while avoiding some of these issues, can be resource-intensive as each process requires its own memory space.
    Memory errors:
    When working with large datasets or inefficient code that consumes more memory than is available, MemoryError can occur. 
    
    24. How do you drive an exception manually in Python?
    Ans:- 
           To manually trigger an exception in Python, the raise keyword is used, followed by the exception class or an instance of it. Optionally, a custom error message can be included for clarity.
    Python
    
    raise Exception("This is a manually raised exception")
    
    raise ValueError("Invalid input provided")
    
    def process_data(data):
        if not isinstance(data, list):
            raise TypeError("Data must be a list")
        # Further processing
    It's also possible to create custom exception classes inheriting from the base Exception class or its subclasses. This allows for more specific error handling.
    Python
    
    class CustomError(Exception):
        """Custom exception class"""
        pass
    
    raise CustomError("Something went wrong")
    When an exception is raised, it halts the normal program flow and searches for an appropriate except block to handle it. If no matching except block is found, the program terminates with an error message.
    
    raise Exception("This is a manually raised exception")
    
    raise ValueError("Invalid input provided")
    
    def process_data(data):
        if not isinstance(data, list):
            raise TypeError("Data must be a list")
        # Further processing
    
    25. Why is it important to use multithreading in certain applications? 
    Ans:- 
            Multithreading is important in certain Python applications primarily to improve performance and responsiveness. It allows multiple threads to run concurrently within a single process, enabling parallel execution of tasks. This is particularly beneficial in scenarios involving I/O-bound operations, where the program spends a significant amount of time waiting for external resources. By using multithreading, these wait times can be overlapped, leading to faster overall execution.
    However, due to Python's Global Interpreter Lock (GIL), true parallelism for CPU-bound tasks is limited. The GIL allows only one thread to execute Python bytecode at a time, even on multi-core processors. Therefore, multithreading is more effective for I/O-bound tasks and for maintaining responsiveness in applications with user interfaces.
    Common use cases where multithreading is valuable include:
    I/O-bound operations: Downloading files, making network requests, or reading/writing to disk.
    User interface applications: Keeping the UI responsive while performing background tasks.
    Web servers: Handling multiple client requests concurrently.
    Data processing: Processing large datasets in parallel (though multiprocessing might be more suitable for CPU-bound tasks).
    Game development: Managing game logic, rendering, and AI simultaneously.
    Automation: Handling multiple sensor inputs in robotics.

'''

# PRACTICAL #
```

```python
#9. How can you read a file line by line and store its content in a list  in Python?#
def file_to_list(file_path):
    """Reads a file line by line and stores each line in a list.

    Args:
        file_path: The path to the file.

    Returns:
        A list where each element is a line from the file, or None if an error occurs.
    """
    try:
        with open(file_path, 'r') as file:
            lines = file.readlines()
            # Remove trailing newlines from each line
            lines = [line.rstrip('\n') for line in lines]
            return lines
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

# Example usage:
file_path = "my_text_file.txt"
file_content_list = file_to_list(file_path)

if file_content_list:
    print(file_content_list)
```

```python
# 14. Write a program that uses the logging module to log both information and error messages in Python.#
import logging

# Configure the logging module
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    filename='app.log',
    filemode='w'
)

# Function to perform division
def divide(x, y):
    logging.info(f"Dividing {x} by {y}")
    try:
        result = x / y
    except ZeroDivisionError:
        logging.error(f"Attempted to divide {x} by zero")
        return None
    else:
        return result

# Example usage
num1 = 10
num2 = 2
result = divide(num1, num2)
if result is not None:
    logging.info(f"The result of {num1} / {num2} is {result}")

num3 = 5
num4 = 0
result = divide(num3, num4)
if result is None:
    logging.info(f"Division of {num3} by {num4} failed.")
```

```python
#15. Write a Python program that prints the content of a file that handle the case when the file is empty. #
def print_file_content(file_path):
    """
    Prints the content of a file.
    If the file is empty, it prints a message indicating that the file is empty.
    """
    try:
        with open(file_path, 'r') as file:
            content = file.read()
            if not content:
                print("The file is empty.")
            else:
                print(content)
    except FileNotFoundError:
        print(f"Error: File not found: {file_path}")

# Example usage:
file_path = "my_file.txt" 
# Create an empty file for testing
with open(file_path, 'w') as f:
    pass
print("Content of the empty file:")
print_file_content(file_path)

# Write some content to the file
with open(file_path, 'w') as f:
    f.write("This is a test file.\nIt has some content.")

print("\nContent of the file with content:")
print_file_content(file_path)

file_path_not_existent = "not_existent.txt"
print("\nContent of not existent file:")
print_file_content(file_path_not_existent)
```

```python
#17. Write a Python program to create and write a list of numbers to a file one number per line.#
def write_numbers_to_file(filename, numbers):
    """Writes a list of numbers to a file, one number per line.

    Args:
        filename: The name of the file to write to.
        numbers: A list of numbers to write.
    """
    with open(filename, 'w') as file:
        for number in numbers:
            file.write(str(number) + '\n')

# Example usage:
numbers = [1, 2, 3, 4, 5]
filename = 'numbers.txt'
write_numbers_to_file(filename, numbers)

print(f"Numbers have been written to '{filename}'")
```

```python
# 18. How would you implement a basic logging setup that logs to a file with rotation after 1 MB#
import logging
from logging.handlers import RotatingFileHandler
import os

log_file = "app.log"
log_max_size = 1024 * 1024  # 1 MB
log_backup_count = 5

# Create a logger
logger = logging.getLogger(__name__)
logger.setLevel(logging.DEBUG)

# Create a rotating file handler
handler = RotatingFileHandler(log_file, maxBytes=log_max_size, backupCount=log_backup_count)

# Create a logging format
formatter = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
handler.setFormatter(formatter)

# Add the handler to the logger
logger.addHandler(handler)

# Example usage
logger.debug("This is a debug message")
logger.info("This is an info message")
logger.warning("This is a warning message")
logger.error("This is an error message")
logger.critical("This is a critical message")

# Check file size
if os.path.exists(log_file):
    file_size = os.path.getsize(log_file)
    print(f"The size of {log_file} is: {file_size} bytes")
```

```python
#20. How would you open a file and read its content using a context manager in Python.#
def read_file_with_context_manager(filename):
    """
    Opens a file in read mode using a context manager and returns its content.

    Args:
        filename (str): The name of the file to be read.

    Returns:
        str: The content of the file, or None if an error occurs.
    """
    try:
        with open(filename, 'r') as file:
            content = file.read()
        return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# Example usage:
filename = 'example.txt'

# Create a dummy file for testing
with open(filename, 'w') as f:
    f.write("This is a test file.\nIt has multiple lines.")

content = read_file_with_context_manager(filename)

if content:
    print(content)
```

```python
#21. Write a Python program that read the file and prints the number of occurrence of a specific word#
def count_word_occurrences(filepath, word):
    """
    Reads a file and counts the number of occurrences of a specific word.

    Args:
        filepath (str): The path to the file.
        word (str): The word to search for.

    Returns:
        int: The number of occurrences of the word in the file.
             Returns -1 if the file cannot be opened.
    """
    try:
      with open(filepath, 'r') as file:
        text = file.read()
        words = text.lower().split()
        return words.count(word.lower())
    except FileNotFoundError:
        print(f"Error: File not found at path: {filepath}")
        return -1
    except Exception as e:
        print(f"An error occurred: {e}")
        return -1
# Example usage:
filepath = 'my_text_file.txt'
word = 'example'

# Create a dummy file for testing
with open(filepath, 'w') as f:
    f.write("This is an example file. It contains example text. example example")

count = count_word_occurrences(filepath, word)

if count != -1:
    print(f"The word '{word}' appears {count} times in the file.")

# Clean up the dummy file
import os
os.remove(filepath)
```

```python
#22. How can you check if a file if empty before attempting to read its contents?#
import os

def is_file_empty(file_path):
    """
    Checks if a file is empty.

    Args:
        file_path (str): The path to the file.

    Returns:
        bool: True if the file is empty, False otherwise.
    """
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")
    return os.path.getsize(file_path) == 0

# Example usage:
file_path = "my_file.txt"

# Create an empty file for testing
with open(file_path, 'w') as f:
    pass

if is_file_empty(file_path):
    print(f"The file '{file_path}' is empty.")
else:
    print(f"The file '{file_path}' is not empty.")

# Clean up the created file
os.remove(file_path)
```
